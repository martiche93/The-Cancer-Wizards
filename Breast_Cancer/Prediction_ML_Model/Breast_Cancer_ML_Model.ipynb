```python
# import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read in csv & create df
df = pd.read_csv("breast_cancer_prediction.csv")
df
```

```python
# count rows and columns
df.shape
```

```python
# find & clean null columns
df.isna().sum()
```

```python
# only 1 column dropped
df.dropna(axis = 1, inplace = True)
df
```

```python
# assign X(data) & y(target)
# y = diagnosis, must be dropped from the X df
X = df.drop("diagnosis", axis = 1)
y = df["diagnosis"]
print(X.shape, y.shape)
```

```python
# split the data into training & testing using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
X_train
```

**Logistic Regression**

```python
# create logistic regression model
# scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# create a logistic regression model
from sklearn.linear_model import LogisticRegression

# clf = classifier
clf = LogisticRegression(solver = "liblinear")
```

```python
# fit the model to the data
clf.fit(X_train, y_train)
```

```python
# validate the model using the test data
# print the r2 score for the test data
print(f"Training Data Score: {clf.score(X_train, y_train)}")
print(f"Testing Data Score: {clf.score(X_test, y_test)}")
```

```python
# make predictions using the X_test & y_test data
# print 15 predictions vs actual diagnosis
y_pred = clf.predict(X_test)
type(y_pred)
```

**Encoding the Diagnosis Column**

```python
# read in csv & create df
df = pd.read_csv("breast_cancer_prediction.csv")
df.dropna(axis = 1, inplace = True)
df
```

```python
# encode the diagnosis values with label encoding
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,1] = labelencoder.fit_transform(df.iloc[:,1].values)
df
```

```python
df.to_csv("label_encoded_breast_cancer_prediction.csv")
```

```python
target = df["diagnosis"]
target_names = ["malignant", "benign"]
```

```python
data = df.drop("diagnosis", axis = 1)
feature_names = data.columns
data
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 42)
```

```python
# Create a Decision Tree Classifier
from sklearn import tree

dt = tree.DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt.score(X_test, y_test)
```

```python
# Fit the classifier to the data
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 800)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Create, fit, and score a Random Forest Classifier
sorted(zip(rf.feature_importances_, feature_names), reverse = True)
```
