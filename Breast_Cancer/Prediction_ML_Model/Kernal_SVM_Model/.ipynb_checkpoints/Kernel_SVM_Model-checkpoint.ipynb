```python
# import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read in csv & create df
df = pd.read_csv("../Resources/label_encoded_breast_cancer_prediction.csv")
df
```

```python
df.drop(["Unnamed: 0"], axis=1)
```

```python
# assign X(data) & y(target)
# y = diagnosis, must be dropped from the X df
X = df.drop("diagnosis", axis = 1)
y = df["diagnosis"]
print(X.shape, y.shape)
```

```python
# split the data into training & testing using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = "rbf")
model.fit(X_train, y_train)
```

```python
# validate the model using the test data
# print the r2 score for the test data
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
#make use of trained model to make predictions on test data
y_predictions = model.predict(X_test)
type(y_predictions)
```

```python
#plot confusion matrix
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix

data = confusion_matrix(y_test, y_predictions)
df_cm = pd.DataFrame(data, columns = np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = "Actual"
df_cm.columns.name = "Predicted"
plt.figure(figsize = (10, 7))
sns.set(font_scale = 1.4)
confusion_matrix_plot = sns.heatmap(df_cm, cmap = "Blues", annot = True, annot_kws = {"size": 16})
confusion_matrix_plot.figure.savefig("Kernel_SVM_Model_confusion_matrix_plot.png")
```

## **REMEMBER: Benign = 0 & Malignant = 1**
**The above confusion matrix visually shows how our model did:**

86 True Positives = True Benign Tumors

3 False Positives = Predicted to be Malignant Tumors, but actually Benign Tumors

2 False Negatives = Predicted to be Benign Tumors, but actually Malignant Tumors

52 True Negatives = True Malignant Tumors

```python
# Model Accuracy (same as above for testing data)
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_predictions)
print(report)
```

```python
#create dataframe to compare our model's prediction performance AFTER hypertuning
prediction_df = pd.DataFrame(
    {
        "Predictive Diagnosis": y_predictions, 
        "Actual Diagnosis": y_test
    })
prediction_df
```

```python
#save predicted df to CSV file
prediction_df.to_csv("../Prediction_Output_CSVs/Kernel_SVM_Model_Prediction.csv")
```
