```python
# import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read in csv & create df
df = pd.read_csv("label_encoded_breast_cancer_prediction.csv")
df
```

```python
df.drop(["Unnamed: 0"], axis=1)
```

```python
# assign X(data) & y(target)
# y = diagnosis, must be dropped from the X df
X = df.drop("diagnosis", axis = 1)
y = df["diagnosis"]
print(X.shape, y.shape)
```

```python
# split the data into training & testing using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

**Support Vector Machine Linear Classifier**

```python
# scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = "linear")
model.fit(X_train, y_train)
```

```python
# validate the model using the test data
# print the r2 score for the test data
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
#make use of trained model to make predictions on test data
y_predictions = model.predict(X_test)
type(y_predictions)
```

```python
y_predictions
```

```python
#plot confusion matrix
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix

data = confusion_matrix(y_test, y_predictions)
df_cm = pd.DataFrame(data, columns = np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = "Actual"
df_cm.columns.name = "Predicted"
plt.figure(figsize = (10, 7))
sns.set(font_scale = 1.4)
confusion_matrix_plot = sns.heatmap(df_cm, cmap = "Blues", annot = True, annot_kws = {"size": 16})
confusion_matrix_plot.figure.savefig("SVC_MODEL_confusion_matrix_plot.png")
```

## **REMEMBER: Benign = 0 & Malignant = 1**
**The above confusion matrix visually shows how our model did:**

85 True Positives = True Benign Tumors

4 False Positives = Predicted to be Malignant Tumors, but actually Benign Tumors

2 False Negatives = Predicted to be Benign Tumors, but actually Malignant Tumors

52 True Negatives = True Malignant Tumors

```python
# Model Accuracy (same as above for testing data)
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_predictions)
print(report)
```

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
# Try adjusting `C` with values of 1, 5, and 10. Adjust `gamma` using .0001, 0.001, and 0.01
from sklearn.model_selection import GridSearchCV

param_grid = {"C": [1, 5, 10],
              "gamma": [0.001, 0.001, 0.01]
             }
grid = GridSearchCV(model, param_grid, verbose = 3)
```

```python
# Fit the model using the grid search estimator. 
# This will take the SVC model and try each combination of parameters
grid.fit(X_train, y_train)
```

```python
# List the best parameters for this dataset
print(grid.best_params_)
```

```python
# List the best score
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
y_pred = grid.predict(X_test)
```

```python
#plot confusion matrix using hypertuned model
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix

data = confusion_matrix(y_test, y_pred)
df_cm = pd.DataFrame(data, columns = np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = "Actual"
df_cm.columns.name = "Predicted"
plt.figure(figsize = (10, 7))
sns.set(font_scale = 1.4)
confusion_matrix_plot = sns.heatmap(df_cm, cmap = "Blues", annot = True, annot_kws = {"size": 16})
confusion_matrix_plot.figure.savefig("hypertuned_SVC_MODEL_confusion_matrix_plot.png")
```
