```python
# import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read in csv & create df
df = pd.read_csv("label_encoded_breast_cancer_prediction.csv")
df
```

```python
df.drop(["Unnamed: 0"], axis=1)
```

```python
# assign X(data) & y(target)
# y = diagnosis, must be dropped from the X df
X = df.drop("diagnosis", axis = 1)
y = df["diagnosis"]
print(X.shape, y.shape)
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Create a Decision Tree Classifier
from sklearn import tree

dt = tree.DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt.score(X_test, y_test)
```

```python
# Fit the classifier to the data
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 500)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
#make use of trained model to make predictions on test data
y_predictions = rf.predict(X_test)
type(y_predictions)
```

```python
#plot confusion matrix
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix

data = confusion_matrix(y_test, y_predictions)
df_cm = pd.DataFrame(data, columns = np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = "Actual"
df_cm.columns.name = "Predicted"
plt.figure(figsize = (10, 7))
sns.set(font_scale = 1.4)
confusion_matrix_plot = sns.heatmap(df_cm, cmap = "Blues", annot = True, annot_kws = {"size": 16})
confusion_matrix_plot.figure.savefig("RANDOM_FOREST_MODEL_confusion_matrix_plot.png")
```

## **REMEMBER: Benign = 0 & Malignant = 1**
**The above confusion matrix visually shows how our model did:**

88 True Positives = True Benign Tumors

1 False Positives = Predicted to be Malignant Tumors, but actually Benign Tumors

3 False Negatives = Predicted to be Benign Tumors, but actually Malignant Tumors

51 True Negatives = True Malignant Tumors

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_predictions)
print(report)
```
