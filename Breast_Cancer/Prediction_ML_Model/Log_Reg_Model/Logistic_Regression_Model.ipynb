# Label Encoding

```python
#get Datatypes of each column in our dataset
df.dtypes
```

```python
#encode the diagnosis values
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,1] = labelencoder.fit_transform(df.iloc[:,1].values)
```

# Split Dataset & Feature Scaling

```python
#splitting the dataset into independent and dependent datasets 
#diagnosis column = dependent variable = in position #1
X = df.iloc[:,2:].values #all values after position #1
y = df.iloc[:,1].values #only value in position #1
```

```python
#splitting datasets into training(75%) and testing(25%)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
#scaling the data (feature scaling) to bring into the same range
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#print data
X_train
```

# Build a Logistic Regression Model

```python
#build a logistic regression classifier
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
#print the r2 score for the training & test data
print(f"Training Data Score: {clf.score(X_train, y_train)}")
print(f"Testing Data Score: {clf.score(X_test, y_test)}")
```

```python
#make use of trained model to make predictions on test data
y_predictions = clf.predict(X_test)
type(y_predictions)
```

```python
y_predictions
```

```python
#compare the predicted vs actual
print(f"First 20 Prediction Diagnoses {y_predictions[:20]}")
print(f"First 20 Actual Diagnoses {y_test[:20]}")
```

```python
#create dataframe to compare our model's prediction performance
prediction_df = pd.DataFrame(
    {
        "Predictive Diagnosis": y_predictions, 
        "Actual Diagnosis": y_test
    })
prediction_df
```

```python
#save predicted df to CSV file
prediction_df.to_csv("../Prediction_Output/Log_Reg_Prediction.csv")
```

# Performance Evaluation

TP = true positive

FP = false positive

TN = true negative

FN = false negative

![](confusion_matrix.jpg)

```python
#plot confusion matrix
import numpy as np
from sklearn.metrics import confusion_matrix

data = confusion_matrix(y_test, y_predictions)
df_cm = pd.DataFrame(data, columns = np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = "Actual"
df_cm.columns.name = "Predicted"
plt.figure(figsize = (10, 7))
sns.set(font_scale = 1.4)
confusion_matrix_plot = sns.heatmap(df_cm, cmap = "Blues", annot = True, annot_kws = {"size": 16})
confusion_matrix_plot.figure.savefig("LOG_REG_MODEL_confusion_matrix_plot.png")
```

## **REMEMBER: Benign = 0 & Malignant = 1**
**The above confusion matrix visually shows how our model did:**

89 True Positives = True Benign Tumors

0 False Positives = Predicted to be Malignant Tumors, but actually Benign Tumors

2 False Negatives = Predicted to be Benign Tumors, but actually Malignant Tumors

52 True Negatives = True Malignant Tumors

```python
#get accuracy score for model
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_predictions))
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_predictions)
print(report)
```
